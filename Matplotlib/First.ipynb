```python
import matplotlib.pyplot as plt

x=["MONDAY","TUEDAY","WEDNESDAY","FRIDAY","TUURSDAY","FRIDAY"] #x-axis
y=[10,17,5,12,2,14] #y-axis

plt.plot(x,y)
plt.title("bakery cake sale")

plt.xlabel('Day of the week')
plt.ylabel("cake per day sale out")
plt.legend()
plt.show()
```
